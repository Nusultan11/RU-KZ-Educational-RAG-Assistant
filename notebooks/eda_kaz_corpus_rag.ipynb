```python
from datasets import load_dataset
import pandas as pd
```

```python
kz_dataset = load_dataset(
    "shyngys879/Kazakh-Wiki-RAG-Dataset"
)
```

```python
base_url = "hf://datasets/shyngys879/Kazakh-Wiki-RAG-Dataset"

kz_train = pd.read_csv(f"{base_url}/train.csv")
kz_pairs = pd.read_csv(f"{base_url}/train_pairs.csv")
kz_triplets = pd.read_csv(f"{base_url}/train_triplets.csv")

print("train:", kz_train.shape)
print("pairs:", kz_pairs.shape)
print("triplets:", kz_triplets.shape)
```

```python
print("TRAIN")
print(kz_train.columns.tolist())

print("\nPAIRS")
print(kz_pairs.columns.tolist())

print("\nTRIPLETS")
print(kz_triplets.columns.tolist())
```

```python
kz_triplets["anchor_len"] = kz_triplets["anchor"].str.len()
kz_triplets["positive_len"] = kz_triplets["positive"].str.len()
kz_triplets["negative_len"] = kz_triplets["negative"].str.len()

kz_triplets[
    ["anchor_len", "positive_len", "negative_len"]
].describe(
    percentiles=[0.5, 0.75, 0.9, 0.95, 0.99]
)
```
